```python
import pandas as pd
from datetime import datetime as dt
```

```python
data = pd.read_csv('AAPL_twitter_cleaned.csv')
data.rename(columns={'Unnamed: 0' :'sentiment_score'}, inplace=True )
```

```python
groupbydate = data.groupby('date').mean()
```

```python
def weighted_weekends(data):
    weighted = 0
    new_df = []
    for i in range(len(data)):
        day = dt.strptime(groupbydate.index[i], "%Y-%m-%d").weekday()
        if (day == 5): # saturday
            weighted += data.iloc[i][0] * 0.2
        elif (day == 6): # sunday
            weighted += data.iloc[i][0] * 0.3
        elif (day == 0): # monday
            weighted += data.iloc[i][0] * 0.5
            new_df.append([groupbydate.index[i], weighted])
            weighted = 0
        else:
            new_df.append([groupbydate.index[i], groupbydate.iloc[i][0]])
            
    return new_df
```

```python
weighted_sentiment = pd.DataFrame(weighted_weekends(groupbydate), columns = ['date', 'sentiment_score'])
weighted_sentiment
```
