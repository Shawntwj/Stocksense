```python
combined_csv = combined_csv[["Ticker", "Title", "Article", "Summary", "Link" ,"Date"]]
```

```python
combined_csv = combined_csv.dropna()
combined_csv['date'] = pd.to_datetime(combined_csv['Date'], errors='coerce', utc=True)
combined_csv["date"] = pd.to_datetime(combined_csv['date']).dt.tz_localize(None)
combined_csv["date"] = pd.to_datetime(combined_csv['date']).dt.date
```

```python
combined_csv["content"] = combined_csv["Article"] +  " " +combined_csv["Summary"]
```

```python
import re
import nltk
from nltk import word_tokenize
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

def remove_hashtag_mentions_urls(text):
    return re.sub(r"(?:\@|\#|https?\://)\S+", "", text)

def remove_emoji(text):
    emoji_pattern = re.compile("["
    u"\U0001F600-\U0001F64F"  # emoticons
    u"\U0001F300-\U0001F5FF"  # symbols & pictographs
    u"\U0001F680-\U0001F6FF"  # transport & map symbols
    u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
    u"\U00002702-\U000027B0"
    u"\U000024C2-\U0001F251"
    "]+", flags=re.UNICODE)

    return emoji_pattern.sub(r'', text)

def tokenization(text):
    word_tokenizer = RegexpTokenizer(r'[-\'\w]+')
    tokenized_text = word_tokenizer.tokenize(text)
    return tokenized_text
```

```python
import numpy as np
def clean_news(df):

    ticketdict = {'aapl':"apple", 'amzn':'amazon', 'ark':'arkk', 'ater':'aterian', 'goog':'google', 'msft':'microsoft', 'rui':'russell', 'tsla':'tesla', 'vxrt':'vaxart'}
    key_list = list(ticketdict)
    values = ticketdict.values()
    values_list = list(values)

    stop_words = stopwords.words('english')   
    stop_words.extend(["stocks","share","shares","market","let","gt","still","got","really","https","post","subreddit","moderator","think","company","companies","discussion","moderators"])
    stop_words.extend(key_list)
    stop_words.extend(values_list)
    
    
    for index,row in df.iterrows():
        
        individual_text = row["content"]
        individual_text = word_tokenize(individual_text)
        individual_text_lower = [w.lower() for w in individual_text]
        individual_text_only = [w for w in individual_text_lower if re.search('^[a-z]+$',w)]
        individual_text_no_emoji =  [w for w in individual_text_only if remove_emoji(w)]
        text_stopremoved = [w for w in individual_text_no_emoji if w not in stop_words]
        individual_text_sentence = ' '.join(text_stopremoved)
        
        if individual_text_sentence == "":
            df.at[index,'cleaned_text'] = np.nan
            
        else:
            df.at[index,'cleaned_text'] = individual_text_sentence
        
        
        individual_text = row["Summary"]
        individual_text = word_tokenize(individual_text)
        individual_text_lower = [w.lower() for w in individual_text]
        individual_text_only = [w for w in individual_text_lower if re.search('^[a-z]+$',w)]
        individual_text_no_emoji =  [w for w in individual_text_only if remove_emoji(w)]
        text_stopremoved = [w for w in individual_text_no_emoji if w not in stop_words]
        individual_text_sentence = ' '.join(text_stopremoved)
        df.at[index,'cleaned_summary_text'] = individual_text_sentence
    return df
```

```python
clean_combined_csv = clean_news(combined_csv)
```

```python
import pandas as pd
import flair
flair_sentiment = flair.models.TextClassifier.load('en-sentiment')  # Load model


def flair_senitment(ticketdf):
    for index, row in ticketdf.iterrows():

        try:
            s = flair.data.Sentence(row["cleaned_text"])  # Retrieve title of post

            flair_sentiment.predict(s) 
            sentiment = str(s.labels[0]).split()[0]

            score = str(s.labels[0]).split()[1][1:-1]
            ticketdf.at[index,'flair_sentiment'] = sentiment
            ticketdf.at[index,'flair_score'] = score
        except:
            ticketdf.at[index,'flair_sentiment'] = "neutral"
            ticketdf.at[index,'flair_score'] = 0
            pass

    return ticketdf


        
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

def vader_senitment(ticketdf):
    for index, row in ticketdf.iterrows():
        
        vs = sia.polarity_scores(row["cleaned_text"])
        
        score = vs["compound"]
        del vs['compound']
        sentiment = max(vs, key=vs.get)

        ticketdf.at[index,'vader_sentiment'] = sentiment
        ticketdf.at[index,'vader_score'] = score
    
    return ticketdf

import pandas as pd
from happytransformer import HappyTextClassification 
happy_tc = HappyTextClassification("BERT", "ProsusAI/finbert", num_labels=3)


def finbert_senitment(ticketdf):

    for index, row in ticketdf.iterrows():
        if len(row["cleaned_summary_text"].split()) > 511:
            result = happy_tc.classify_text(" ".join(row["cleaned_summary_text"].split()[:400]))
        else:
            result = happy_tc.classify_text(" ".join(row["cleaned_summary_text"].split()[:400]))
        sentiment = result.label
        score = result.score

        ticketdf.at[index,'finbert_sentiment'] = sentiment
        ticketdf.at[index,'finbert_score'] = score
    
    return ticketdf
```

```python
clean_combined_csv = flair_senitment(clean_combined_csv)
clean_combined_csv = vader_senitment(clean_combined_csv)
clean_combined_csv = finbert_senitment(clean_combined_csv)
```
