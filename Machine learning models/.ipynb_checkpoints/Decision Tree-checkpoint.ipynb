decision tree
1. sentiment score for 24hrs
2. correlation score
3. predicted price for the day
4. yesterday's price

```python
# this is not the machine learning decision tree
# just a series of yes and no


def decision_tree(sentiment, correlation, predicted_price, yesterday_price):
    
    gate = 0
    
    # if today will be more than tmr should buy
    if predicted_price > yesterday_price:
        gate += 1
    else:
        gate -= 1
        
    
    if sentiment >= 0:
        if correlation >= 0.2:
            gate += 1
        if correlation >= 0.4:
            gate += 2
        if correlation <= -0.2:
            gate -= 1
        if correlation <= -0.4:
            gate -= 2
    
    else:
        if correlation <= -0.2:
            gate += 1
        if correlation <= -0.4:
            gate += 2
        if correlation >= 0.2:
            gate -= 1
        if correlation >= 0.4:
            gate -= 2
    
    print(gate)
    if gate >= 2:
        return "Buy"
    elif gate <= -2:
        return "Sell"
    else:
        return "Hold"
```

```python
# test
print(decision_tree(0.4, -0.7, 1.5, 1.0))
```
