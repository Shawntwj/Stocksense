```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
# Creating list to append tweet data to
tweets_list2 = []
queries = [ "Vxrt", "russell 1000", "Ark", "AAPL", "Tesla", "Microsoft", "Google", "Amazon"]

for query in queries:
    # Using TwitterSearchScraper to scrape data and append tweets to list
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(query+' since:2021-01-01 until:2021-09-30').get_items()):
        # min word length = 5, min like = 200, min followers = 50, min retweet = 5
        if len(tweet.content.split())>=5 and tweet.likeCount>=200 and tweet.user.followersCount>=50 and tweet.retweetCount>=5:
            tweets_list2.append([tweet.user.username, tweet.content, tweet.date, tweet.user.followersCount, tweet.replyCount, tweet.retweetCount, tweet.likeCount])
    
    # Creating a dataframe from the tweets list above
    tweets_df2 = pd.DataFrame(tweets_list2, columns=['Username', 'Tweet', 'Datetime', 'Number of Followers', 'Number of Comments', 'Number of Retweets', 'Number of Likes'])
    tweets_df2.head(10)
    tweets_df2.to_csv(query+".csv")
```
