```python
import csv
import glob
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
filenames = glob.glob("*cleaned_sentiment*.csv")

final_df = pd.DataFrame()

for file in filenames:
    df = pd.read_csv(file)
    final_df = pd.concat([final_df, df])

# changing flaire sentiment score to match flaire_sentiment
mask = (final_df['flair_sentiment'] == 'NEGATIVE')
final_df['flair_sentiment_score'][mask] *= -1


final_df.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1', 'Username', 'Tweet', 'cleaned Tweet', 'flair_sentiment', 'vader_sentiment'], inplace = True)


final_df.head()
```
